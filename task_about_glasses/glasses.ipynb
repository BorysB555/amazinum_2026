### Для k стаканів і n поверхів

Далі для задачі на k стаканів і n поверхів я користувався такою формулою: 
F(n,k) = F(n-1, k-1) + F(n-1, k) + 1.

- n — скільки разів ми готові кидати стакан (ходів).
- k — скільки у нас cклянок.
- F(n,k) — максимальна кількість поверхів, яку ми гарантовано можемо перевірити за n ходів і k cклянок, без ризику пропустити «критичний» поверх.

Коли стакан розбивається, ви відразу переходите до випадку з 𝑘 − 1 склянками і n−1 ходами (ліворуч у формулі). Коли стакан не розбивається, відразу переходимо до випадку з k склянкам і n−1 ходами (праворуч). +1 завжди враховує поточну перевірку рівня.
Ця рекурсія працює для будь‑якого k, бо кожен кидок ділить простір можливостей на дві гілки (розбилось / не розбилось), а +1 дає змогу зарахувати саме цей експеримент. 
На основі цієї формули я створив табличку 100х5, де рядки - кількість спроб, стовбці - кількість склянок, а на перетині максимальний поверх який вдасться порахувати цією комбінацією найефективніше.

### Код

```python
import pandas as pd


for attempt in range(1, 5):
    floors = int(input("Введіть кількість поверхів: "))
    glasses = int(input("Введіть кількість стаканів: "))

    table = pd.DataFrame(
        index=range(0, floors + 1),
        columns=range(1, glasses + 1),
        dtype=int
    )

    for k in range(1, glasses + 1):
        table.at[0, k] = 0
    # F(n, 1) = n
    for n in range(1, floors + 1):
        table.at[n, 1] = n

    # Основна рекурсія та пошук мінімального n
    required_trials = None
    for n in range(1, floors + 1):
        for k in range(2, glasses + 1):
            table.at[n, k] = table.at[n - 1, k - 1] + table.at[n - 1, k] + 1

        if required_trials is None and table.at[n, glasses] >= floors:
            required_trials = n

    print(f"\nМінімальна кількість спроб для {floors} поверхів і {glasses} стаканів: {required_trials}\n")
```
