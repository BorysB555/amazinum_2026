```python
import numpy as np
import matplotlib.pyplot as plt

# Клас для збереження інформації про найглибшу річку
class DeepestRiver:
    def __init__(self):
        self.left_index = None  # Ліва межа річки
        self.right_index = None  # Права межа річки
        self.depth = 0  # Глибина річки  # Індекс найнижчої точки річки

# Функція для знаходження найглибшої річки у рельєфі

def find_deepest_river(heights):
    river = DeepestRiver()
   
    # Перевірка зліва направо
    left_bank = (0, heights[0])  # Початкова ліва межа
    current_max_depth = 0  # Поточна максимальна глибина
    current_deepest_point = None  # Індекс найглибшої точки
   
    for i in range(1, len(heights)):
        if heights[i] < left_bank[1]:  # Якщо поточна точка нижча за лівий берег
            depth = left_bank[1] - heights[i]
            if depth > current_max_depth:
                current_max_depth = depth
                current_deepest_point = i  # Запам’ятовуємо найнижчу точку
        elif heights[i] > left_bank[1]:  # Якщо поточна точка вища за лівий берег
            if current_max_depth > river.depth:  # Якщо знайдена річка глибша за попередню
                river.left_index = left_bank[0]
                river.right_index = i
                river.depth = current_max_depth
                river.lowest_point_index = current_deepest_point
            left_bank = (i, heights[i])  # Оновлюємо лівий берег
            current_max_depth = 0  # Скидаємо поточну глибину
   
    # Перевірка справа наліво
    right_bank = (len(heights) - 1, heights[-1])  # Початковий правий берег
    current_max_depth = 0  # Скидаємо поточну максимальну глибину
   
    for i in range(len(heights) - 2, -1, -1):
        if heights[i] < right_bank[1]:  # Якщо точка нижча за правий берег
            depth = right_bank[1] - heights[i]
            if depth > current_max_depth:
                current_max_depth = depth
        elif heights[i] > right_bank[1]:  # Якщо точка вища за правий берег
            if current_max_depth > river.depth:  # Якщо знайдена річка глибша за попередню
                river.left_index = i
                river.right_index = right_bank[0]
                river.depth = current_max_depth
                river.lowest_point_index = current_deepest_point
            right_bank = (i, heights[i])  # Оновлюємо правий берег
            current_max_depth = 0  # Скидаємо поточну глибину
   
    # Якщо річки не знайдено
    if river.depth == 0:
        print("No river found.")
        return None
   
    return river

# Генерація випадкових висот
heights = np.random.randint(0, 100, 20)

# Візуалізація початкових висот
plt.figure(figsize=(8, 5))
plt.plot(heights, label='Heights')
plt.title("Initial Heights Visualization")
plt.show()

# Пошук найглибшої річки
river = find_deepest_river(heights)

# Візуалізація рельєфу з позначеною річкою
plt.figure(figsize=(8, 5))
plt.plot(heights, label='Heights')
if river:
    plt.plot(range(river.left_index, river.right_index + 1),
             heights[river.left_index:river.right_index + 1],
             color='red', linewidth=2.5, label='Deepest River')
    lowest_height = heights[river.lowest_point_index]  # Висота найнижчої точки
    lower_bank_height = min(heights[river.left_index], heights[river.right_index])  # Висота нижчого берега
    plt.axhline(y=lowest_height, color='green', linestyle='dashed')
    plt.axhline(y=lower_bank_height, color='green', linestyle='dashed')

plt.title("Deepest River Highlighted")
plt.show()
```
