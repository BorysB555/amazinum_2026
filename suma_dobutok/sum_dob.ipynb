```python
def find_numbers():
    all_pairs = []
    for i in range(2, 100):
        for j in range(i, 100):
            all_pairs.append((i, j))

#добуток каже "я не знаю цих чисел", це означає, що його добуток можна отримати кількома способами.
    
    #рахую, скільки разів зустрічається кожен добуток
    products_dict = {}
    for a, b in all_pairs:
        prod = a * b
        if prod not in products_dict:
            products_dict[prod] = []
        products_dict[prod].append((a, b))
    
    #знаходжу добутки, які мають більше одного варіанту пар
    duplicate_products = []
    for prod, pairs_list in products_dict.items():
        if len(pairs_list) > 1:
            duplicate_products.append(prod)
            
    #залишаю тільки ті пари, добуток яких є неоднозначним
    step1_pairs = []
    for a, b in all_pairs:
        if (a * b) in duplicate_products:
            step1_pairs.append((a, b))

#сума каже "я знав, що ти їх не знаєш" це означає, що для його суми всі можливі пари дають неоднозначний добуток
    
    #групую всі початкові пари за їхньою сумою
    sums_dict = {}
    for a, b in all_pairs:
        s = a + b
        if s not in sums_dict:
            sums_dict[s] = []
        sums_dict[s].append((a, b))
        
    #відбираю "правильні" суми
    valid_sums = []
    for s, pairs_list in sums_dict.items():
        all_bad = True
        for a, b in pairs_list:
            #якщо хоча б одна пара дає унікальний добуток - ця сума не підходить
            if (a * b) not in duplicate_products:
                all_bad = False
                break
        
        if all_bad:
            valid_sums.append(s)
            
    #фільтрую пари з першого кроку: залишаю ті, що мають "правильну" суму
    step2_pairs = []
    for a, b in step1_pairs:
        if (a + b) in valid_sums:
            step2_pairs.append((a, b))

#добуток каже "тоді тепер я знаю ці числа" серед варіантів, що залишилися, його добуток тепер вказує на одну пару
    
    step3_products_dict = {}
    for a, b in step2_pairs:
        prod = a * b
        if prod not in step3_products_dict:
            step3_products_dict[prod] = []
        step3_products_dict[prod].append((a, b))
        
    #залишаю тільки ті пари, які тепер є єдиним варіантом для свого добутку
    step3_pairs = []
    for prod, pairs_list in step3_products_dict.items():
        if len(pairs_list) == 1:
            step3_pairs.append(pairs_list[0])

#сума каже "тоді і я їх знаю" серед варіантів, що залишилися після 3 кроку, його сума вказує на ОДНУ пару.
    
    step4_sums_dict = {}
    for a, b in step3_pairs:
        s = a + b
        if s not in step4_sums_dict:
            step4_sums_dict[s] = []
        step4_sums_dict[s].append((a, b))
        
    #шукаю фінальну відповідь
    final_answer = []
    for s, pairs_list in step4_sums_dict.items():
        if len(pairs_list) == 1:
            final_answer.append(pairs_list[0])

    return final_answer
```

запуск програми

```python
my_solutions = find_numbers()

for a, b in my_solutions:
    print(f"перше число: {a}, друге число: {b}")
```
